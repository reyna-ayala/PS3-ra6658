```python
import numpy as np
import matplotlib.pyplot as plt


# PROBLEM 1A

# define a grid of coordinates in the complex plane
# ot : omega*Delta(t)
Rmin = -20
Rmax = 2
Imin = -20
Imax = 20

otR, otI = np.meshgrid(np.linspace(Rmin,Rmax,100), np.linspace(Imin,Imax,100))
nR, nI = otI.shape

rho = otR*0

for i in range(nR): #first go through real values
    
    for j in range(nI): #then go through imag values
        
        OT = complex(otR[i,j],otI[i,j])
        
        rho[i,j] = abs(1 + (12*OT)/(OT**2 - 6*OT + 12))

print("Done.")

plt.figure(figsize=(8,7))
plt.contour(otR,otI,rho,np.linspace(0,1,30))
plt.title('DIRK2 Stability', size=18)
plt.xlabel(r'Re($\Omega \Delta t$)', size=14)
plt.ylabel(r'Im($\Omega \Delta t$)', size=14)
plt.colorbar()

plt.show()
```

```python
# PROBLEM 1B, s=1

# define a grid of coordinates in the complex plane
# ot : omega*Delta(t)
Rmin = -10
Rmax = 10
Imin = -10
Imax = 10

otR, otI = np.meshgrid(np.linspace(Rmin,Rmax,100), np.linspace(Imin,Imax,100))
nR, nI = otI.shape

rho_p = otR*0
rho_m = otR*0

rho = otR*0

for i in range(nR): #first go through real values
    
    for j in range(nI): #then go through imag values
        
        OT = complex(otR[i,j],otI[i,j])
        
        rho_p = 2/(3-2*OT) + 1/(6-4*OT)*np.sqrt(4 + 8*OT)
        rho_m = 2/(3-2*OT) - 1/(6-4*OT)*np.sqrt(4 + 8*OT)
        
        rho[i,j] = max(abs(rho_p), abs(rho_m))

print("Done.")

plt.figure(figsize=(8,7))
plt.contour(otR,otI,rho,np.linspace(0,1,100))
plt.colorbar()
plt.title('BDF2 Stability', size=18)
plt.xlabel(r'Re($\Omega \Delta t$)', size=14)
plt.ylabel(r'Im($\Omega \Delta t$)', size=14)

plt.show()
```

```python
# PROBLEM 1B, s=2

# define a grid of coordinates in the complex plane
# ot : omega*Delta(t)
Rmin = -10
Rmax = 10
Imin = -10
Imax = 10

otR, otI = np.meshgrid(np.linspace(Rmin,Rmax,100), np.linspace(Imin,Imax,100))
nR, nI = otI.shape

rho = 1j * np.zeros((nR,nI))

for i in range(nR): #first go through real values
    
    for j in range(nI): #then go through imag values
        
        OT = complex(otR[i,j],otI[i,j])
        
        p = [(1-6/11*OT), -18/11, 9/11, -2/11]
        rho_vals = np.roots(p)
        
        rho[i,j] = max(abs(rho_vals))
        
print("Done.")

plt.figure(figsize=(8,7))
plt.contour(otR,otI,rho, np.linspace(0,1,120))
plt.colorbar()
plt.title('BDF3 Stability', size=18)
plt.xlabel(r'Re($\Omega \Delta t$)', size=14)
plt.ylabel(r'Im($\Omega \Delta t$)', size=14)

plt.show()
```
